```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/bank-additional-full.csv", sep=";")
```

```python
display(df.head())
df.info()
```

```python
display(df.isna().sum())
(df == "unknown").sum()
# se usar o comando comum, mostra que não tem ausentes, mas os ausentes estão como texto "unknown"
# depois é possível ver realmente quantos ausentes se tem no dataset
```

```python
df.duplicated().sum()
# encontrado 12 duplicadatas 
```

```python
df = df.drop_duplicates()
# aqui foi removido as 12 duplicadas
```

```python
df["contatado_antes"] = df["pdays"].apply(lambda x: "nao" if x == 999 else "sim")
# aqui foi criada uma coluna para ver com mais clareza quem foi ou não contatado antes, pdays = 999 significa que não foi
```

```python
df["faixa_etaria"] = pd.cut(
    df["age"],
    bins=[0, 25, 35, 50, 65, 100],
    labels=["ate_25", "26_35", "36_50", "51_65", "acima_65"]
)
# criado uma coluna para transformar as idades numéricas em categorias, vai ajudar a gente na questão de probabilidade depois
```

```python
df.groupby("faixa_etaria")["y"].value_counts(normalize=True)
# foi agrupado e calculado a proporção de yes e no de cada idade
```

```python
df["faixa_campaign"] = pd.cut(
    df["campaign"],
    bins=[0, 1, 3, 6, 60],
    labels=["1", "2a3", "4a6", "mais_6"]
)
# mesmo esquema da faixa_etaria
```
