```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/bank-additional-full.csv", sep=";")
```

```python
display(df.head())
df.info()
```

```python
display(df.isna().sum())
(df == "unknown").sum()
# se usar o comando comum, mostra que não tem ausentes, mas os ausentes estão como texto "unknown"
# depois é possível ver realmente quantos ausentes se tem no dataset
```

```python
df.replace("unknown", np.nan, inplace=True)
df.isna().sum()
# Substitui unknown por np.nan para reconhecimento do pandas
```

```python
for col in df.columns:
    if df[col].mode().empty == False:
        df[col] = df[col].fillna(df[col].mode()[0])

df.isna().sum()
# Substitui os valores ausentes das colunas não vazias pela moda da respectiva coluna
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
# Remove duplicatas
```

```python
df["contatado_antes"] = (df["pdays"] != 999).astype(int)
# Cria uma colona chamada "contatado_antes" considerando o valor 999 na coluna pdays como 0 (nunca foi contatado)
```

```python
df["pdays"].replace(999, 0, inplace=True)
# Substitui o indicador de nunca contatado da coluna "pdays" por 0, para fins de coerência com a coluna "contatado_antes"
```

```python
df = df.replace({'no': 0, 'yes': 1})
# Substitui 'no' por 0 e 'yes' por 1 em todo o DataFrame
```

```python
df["faixa_etaria"] = pd.cut(
    df["age"],
    bins=[0, 25, 35, 50, 65, 100],
    labels=["ate_25", "26_35", "36_50", "51_65", "acima_65"]
)
# criado uma coluna para transformar as idades numéricas em categorias, vai ajudar a gente na questão de probabilidade depois
```

```python
df.groupby("faixa_etaria")["y"].value_counts(normalize=True)
# foi agrupado e calculado a proporção de yes e no de cada idade
```

```python
df["faixa_campaign"] = pd.cut(
    df["campaign"],
    bins=[0, 1, 3, 6, 60],
    labels=["1", "2a3", "4a6", "mais_6"]
)
# mesmo esquema da faixa_etaria
```

```python
df[["age", "campaign", "previous", "duration"]].describe()
# outliers analisados, pois representam situações possíveis no contexto das campanhas
# a variável campaign possui valor máximo bem maior que a mediano, então se tem contatos muito acima do padrão
```

```python
# Conversão para categórica ordinal: education
ordem_educacao = ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree']
df['education'] = pd.Categorical(df['education'], categories=ordem_educacao, ordered=True)
```

```python
# Conversão para categórica ordinal: month e day_of_week
ordem_mes = ['mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
ordem_dia = ['mon', 'tue', 'wed', 'thu', 'fri']

df['month'] = pd.Categorical(df['month'], categories=ordem_mes, ordered=True)
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=ordem_dia, ordered=True)
```

```python
# Conversão para categórica nominal: demais colunas
colunas_nominais = ['job', 'marital', 'default', 'housing', 'loan', 'contact', 'poutcome', 'y', 'contatado_antes']

for col in colunas_nominais:
    df[col] = df[col].astype('category')
```

```python
df_eda = df.copy()
df = df.drop(columns=['duration'])
# A coluna 'duration' causaria conflito no uso dos algoritmos de machine learning
# Mas a coluna será mantida para usar na EDA, para entender o comportamento das chamadas
```

```python
df.to_parquet('dados_limpos.parquet', index = False)
# Arquivo sem o "duration"
df_eda.to_parquet('dados_eda.parquet', index = False)
# Arquivo com o "duration" para o EDA
```
