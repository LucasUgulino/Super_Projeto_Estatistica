```python
import pandas as pd

# Carregando os dados
df = pd.read_parquet('../data/processed/dados_limpos.parquet')

# 1. Definindo as Features (X)
features = [
    'job', 'marital', 'education', 'default', 'housing', 'loan', 
    'contact', 'month', 'day_of_week', 'poutcome', 'faixa_etaria', 
    'faixa_campaign', 'contatado_antes'
]
df_bayes = df[features + ['y']].copy()
df_bayes.head()
```

```python
# 2. Calculando P(C) - Prior
# Qual a probabilidade global de "yes" e "no"?
prior = df_bayes['y'].value_counts(normalize=True).to_dict()
print("Prior P(C):")
print(prior)
```

```python
# 3. Calculando P(X|C) - Likelihood (Verossimilhança)
# Para cada feature, calculamos a proporção de cada categoria dado que y='yes' e dado que y='no'
likelihood = {}

for col in features:
    prop = df_bayes.groupby('y')[col].value_counts(normalize=True).unstack().fillna(0).to_dict('index')
    likelihood[col] = prop

print("Exemplo de Likelihood para 'faixa_etaria' dado 'yes':")
print(likelihood['faixa_etaria'][1])
```
