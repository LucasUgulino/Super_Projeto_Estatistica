```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Carregando os dados
df = pd.read_parquet('dados_limpos.parquet')

# Transformando a variável alvo 'y' em 1 e 0
df['y'] = (df['y'] == 'yes').astype(int)
df.head()
```

```python
# 1. Encoding (One-Hot Encoding)
# get_dummies aplica o encoding nas colunas categóricas (texto) automaticamente.
# drop_first=True evita a armadilha das variáveis dummy (multicolinearidade perfeita)
X = df.drop(columns=['y'])
y = df['y']

X_encoded = pd.get_dummies(X, drop_first=True)
print(f"Número de features após o encoding: {X_encoded.shape[1]}")
```

```python
# 2. Train-Test Split (80% treino, 20% teste)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

print(f"Tamanho do treino: {X_train.shape[0]} amostras")
print(f"Tamanho do teste: {X_test.shape[0]} amostras")
```

```python
# 3. Escalonamento (Scaling)
# Identificando as variáveis originais que são numéricas para aplicar o scaler
colunas_numericas = df.select_dtypes(include=['number']).drop(columns=['y']).columns.tolist()

scaler = StandardScaler()

# O fit e transform no treino
X_train[colunas_numericas] = scaler.fit_transform(X_train[colunas_numericas])

# Apenas transform no teste (evitar data leakage)
X_test[colunas_numericas] = scaler.transform(X_test[colunas_numericas])

X_train.head()
```
